```python
import pandas as pd
```

```python
us_babies = pd.read_csv("us_baby_names.csv")
```

```python
us_babies_1998 = us_babies.loc[us_babies['Year'] == 1998, :]
```

```python
sorted_1998 = us_babies_1998.sort_values('Count', ascending = False)
```

```python
olivia  = us_babies.loc[us_babies['Name'] == 'Olivia', :]
olivia.sort_values('Count', ascending = False)
```

```python
olivia  = olivia.loc[olivia['Gender'] == 'F', :]
olivia  = olivia.loc[olivia['Year'] > 1990, :]
olivia
```

```python
state_names = pd.read_csv('state_baby_names.csv')
```

```python
girls98 = sorted_1998.loc[sorted_1998['Gender']=='F',:]
girls98.iloc[0:25]
```

```python
def popular(s):
    return s.iloc[0]
```

```python
liv_state = state_names.loc[state_names['Name'] == 'Olivia', :]
liv_grouped = liv_state.sort_values('Count', ascending = False).groupby(['Year']).agg(popular)
liv_state = liv_state.loc[liv_state['Gender'] == 'F', :]
liv_state
```
